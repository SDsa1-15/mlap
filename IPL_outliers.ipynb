```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SDsa1-15/mlap/main/Data/IPL%20IMB381IPL2013.csv')
df.head()
```

```python
df.shape
```

```python
min_th = df['SOLD PRICE'].quantile(0.05)
min_th
```

```python
max_th = df['SOLD PRICE'].quantile(0.95)
max_th
```

```python
df[df['SOLD PRICE']>max_th]
```

```python
df[df['SOLD PRICE']<min_th]
```

```python
df[(df['SOLD PRICE']>min_th) & (df['SOLD PRICE']<max_th)]
```
